```python
import requests
from bs4 import BeautifulSoup

# URL for MoneyControl's top gainers page
url = "https://www.moneycontrol.com/markets/top-gainers/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.moneycontrol.com/",
}

# Send request to the URL
response = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table or section with the required information
gainers_table = soup.find('table', {'class': 'mctable1'})

# Extract gainers data
gainers_rows = gainers_table.find_all('tr')[1:]  # Skip the header row

print("Top Gainers from MoneyControl:")
for row in gainers_rows:
    cols = row.find_all('td')
    symbol = cols[0].text.strip()
    price = cols[1].text.strip()
    percent_change = cols[2].text.strip()
    print(f"{symbol} - {price} ({percent_change})")
```

```python
from nsetools import Nse
import pandas as pd # Optional: for better display

try:
    nse = Nse()
    print("Fetching NSE data...") # Add a print statement to show progress

    # Get top gainers (Defaults to NIFTY 50, but you can specify others)
    # Valid indices might include: "NIFTY 50", "NIFTY BANK", "NIFTY NEXT 50", "FNO", "ALL", etc.
    # Refer to nsetools documentation for the exact list.
    gainers = nse.get_top_gainers()
    losers = nse.get_top_losers()

    print("\n--- NSE Top Gainers ---")
    if gainers:
         # Optional: Convert to Pandas DataFrame for nice formatting
        gainers_df = pd.DataFrame(gainers)
        # Select and display relevant columns (check available keys from 'gainers[0]')
        print(gainers_df[['symbol', 'series', 'ltp', 'previousPrice', 'netPrice', 'perChange']].head())
    else:
        print("Could not retrieve gainers data or no gainers found.")


    print("\n--- NSE Top Losers ---")
    if losers:
        # Optional: Convert to Pandas DataFrame
        losers_df = pd.DataFrame(losers)
         # Select and display relevant columns (check available keys from 'losers[0]')
        print(losers_df[['symbol', 'series', 'ltp', 'previousPrice', 'netPrice', 'perChange']].head())
    else:
        print("Could not retrieve losers data or no losers found.")

except Exception as e:
    print(f"An error occurred while fetching NSE data: {e}")
    print("This might be due to network issues, changes in the NSE website structure, or the library needing an update.")
    print("Consider checking the nsetools library documentation or GitHub page for issues.")
```

```python
import httpx

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "application/json",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.nseindia.com/",
    "Connection": "keep-alive",
}

base_url = "https://www.nseindia.com"
api_url = "https://www.nseindia.com/api/live-analysis-variations?index=gainers"

with httpx.Client(headers=headers, timeout=10) as client:
    # Step 1: Visit NSE homepage to get cookies set
    client.get(base_url)

    # Step 2: Now hit the API
    response = client.get(api_url)
    response.raise_for_status()
    data = response.json()
    print(data)
```

```python
import httpx

base_url = "https://www.nseindia.com"
api_url = "https://www.nseindia.com/api/live-analysis-variations?index=gainers"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 " \
                  "(KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36",
    "Accept": "application/json",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.nseindia.com/",
    "Connection": "keep-alive",
    "Accept-Encoding": "gzip, deflate, br",
}

# Use a persistent client with cookie handling
with httpx.Client(headers=headers, timeout=10, follow_redirects=True) as client:
    # Initial GET to get cookies (this is critical)
    home_response = client.get(base_url)
    if home_response.status_code == 200:
        response = client.get(api_url)
        response.raise_for_status()
        data = response.json()
        print(data)
    else:
        print("Failed to establish session with NSE.")
```
